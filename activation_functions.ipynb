```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def sig(x):
  return 1/(1+np.exp(-x))

def dsig(x):
  return sig(x)*(1-sig(x))

x_data=np.linspace(-6,6,100)
y_data=sig(x_data)
dy_data=dsig(x_data)

plt.plot(x_data,y_data,x_data,dy_data)
plt.title("sigmoid")
plt.legend('f(x)','f\'(x)')

plt.grid()
plt.show()
```

```python
def relu(x):
  temp=[max(0,value) for value in x]
  return np.array(temp,dtype=float)

def drelu(x):
  temp=[1 if value>0 else 0 for value in x]
  return np.array(temp,dtype=float)

x_data=np.linspace(-6,6,100)
y_data=relu(x_data)
dy_data=drelu(x_data)

plt.plot(x_data,y_data,x_data,dy_data)
plt.title("Relu")
plt.legend('f(x)','f\'(x)')

plt.grid()
plt.show()
```

```python
def hyp(x):
  return(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def dhyp(x):
  return 1-hyp(x)*hyp(x)

x_data=np.linspace(-6,6,100)
y_data=hyp(x_data)
dy_data=dhyp(x_data)

plt.plot(x_data,y_data,x_data,dy_data)
plt.title("Hyperbolic")
plt.legend('f(x)','f\'(x)')

plt.grid()
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,4))

def leakyrelu(arr):
  lst=[]
  for i in arr:
    if i>=0:
      lst.append(i)
    else:
      lst.append(0.01*i)
  return lst

arr=np.linspace(-5,5)
plt.plot(arr,leakyrelu(arr))
plt.title(leakyrelu)
plt.show()
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x))
x=np.linspace(-5,5)
plt.plot(x,softmax(x))
plt.show()
```
